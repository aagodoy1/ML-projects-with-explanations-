### 2) Import the data

```python
df = pd.read_csv('shampoo.csv')
df.columns=["Month","Sales"]
df.set_index('Month',inplace=True)
print(df.head())
print(df.describe())
print(df.info())
```

As we can see there are no null values and we have 36 measures

### 3) View the behaviour of the sales

```python
rcParams['figure.figsize'] = 9, 3 ## set the size of the figure
df.plot() # define de plot
plt.title('Shampoo Sales')
plt.xlabel('Date')
plt.ylabel('Amount sold')
plt.show() # show
```

As we can see, the behaviour looks non stationary, therefore using the ARIMA model could be a good place to start
